```python
import numpy as np
import pandas as pd
file_path = 'C:/Users/User/Desktop/AICourse/dataset.csv'
data = pd.read_csv(file_path)  
df=pd.DataFrame(data) 
```

```python
#Q1
df['density']=df['2022']/df['Area (km²)']
print(df)
```

```python
#Q2 
df2=pd.DataFrame(data) 
df2['mean']=(df2['2022']+df2['2020']+df2['2015']+df2['2010']+df2['2000']+df2['1990']+df2['1980']+df2['1970'])/8
print(df2)
```

```python
#Q3 
df3=pd.DataFrame(data) 
years=['2022','2020','2015','2010','2000','1990','1980','1970']
for year in years:
   df3[f'percentage[{year}]']=(df3[year] /df3[year].sum())*100
print(df3)
```

```python
#Q4
 
df4=pd.DataFrame(data) 
df4['growth']=df['2022']/df['2020']
print(df4)
```

```python
#Q5
 
df5=pd.DataFrame(data) 
df5_sorted=df5.sort_values('2022',ascending=False)
print(df5_sorted)
```

```python
#Q6
import random
df6=pd.DataFrame(data) 
unique_numbers=random.sample(range(0,235),234)

df6['rank']=unique_numbers
print(df6)    
```

```python
#Q7
df7=pd.DataFrame(data)

df7['binary population class']=np.where(df7['2022']>300_000,'more than 300','less than 300')
df7_sorted = df7.sort_values('2022', ascending=False)
print(df7_sorted)
```

```python
#Q8

df8=pd.DataFrame(data)
df8['percentage']=(df8['2022'])/(df8['2022'].sum())*100

conditions=[
df8['percentage']<2,
 (df8['percentage']>=2)&(df8['percentage']<=5),
df8['percentage']>5   
]
choices=[0,1,'one']

df8['population class'] = np.select(conditions, choices, default=0)
df8_sorted = df8.sort_values('percentage', ascending=False)


df8_sorted
```

```python
#Q9

df9=pd.DataFrame(data)
df9['growth status']=np.where(df9['2022']>df9['1970'],0,1)

print(df9)
```
