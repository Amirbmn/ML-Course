```python
#generalized for any count of labels
from math import sqrt
import matplotlib.pyplot as plt
from collections import defaultdict

def euclidean_distance(point1,point2):
    x1,y1=point1
    x2,y2=point2
    return sqrt((x1-x2)**2+(y1-y2)**2) 



data=[
[(1,2),'A'],
[(4,9),'B'],
[(6,7),'B'],
[(5,4),'C'],
[(2,3),'A'],
[(8,0),'B'],
[(7,6),'D'],
[(8,4),'A'],
[(8,4),'C'],
[(8,4),'D'],
[(8,4),'B']
]



def knn_classify(k,new_point,all_points):    
    sorted_all_points=sorted(all_points,key=lambda item :euclidean_distance(item[0],new_point))
    
    classes_count=defaultdict(int)
    neighbors=[]
    for i in range(k):
        neighbors.append(sorted_all_points[i])
    
        

    for point,label in neighbors:
        classes_count[label]+=1


    return max(classes_count,key=classes_count.get) 

new_point=(9,10)    

print(knn_classify(3,new_point,data))
```

```python
#handle just 2 classes
new_x,new_y=new_point
# Separate data by label
x_A = [x for (x,label)in data if label =='A']
x_B = [x for (x, label) in data if label == 'B']

# Extract x and y values for each label
x_A_vals = [point[0] for point in x_A]
y_A_vals = [point[1] for point in x_A]

x_B_vals = [point[0] for point in x_B]
y_B_vals = [point[1] for point in x_B]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x_A_vals, y_A_vals, color='blue', label='Class A')
plt.scatter(x_B_vals, y_B_vals, color='red', label='Class B')
plt.scatter(new_x,new_y,color='green')
# Annotate
plt.title('2D Data Points by Class')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
