```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.normal(loc=0,scale=1,size=100).tolist()

data.append(15)
data.append(20)
data.append(30)
data.append(-15)
data.append(-20)
data.append(-30)

def z_score(data):
    mean=np.mean(data)
    std_dev=np.std(data)
    return ((data-mean)/std_dev).round(2)

threshold=2
indexes=[]
result=z_score(data).tolist()
z_score_outliers=[]

for element in result:
    if(abs(element)>threshold):
        indexes.append(result.index(element))


for index in indexes:
    z_score_outliers.append(data[index])  


#IQR

IQR_outliers=[]
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

for element in data:
    if(element>lower and element<upper):
        continue
    else:IQR_outliers.append(element)


plt.scatter(data, np.zeros_like(data), s=5)
plt.scatter(z_score_outliers, np.zeros_like(z_score_outliers), s=5,color='red')
plt.yticks([])  
plt.title('Z-score method')
plt.show()



plt.scatter(data, np.zeros_like(data), s=5)
plt.scatter(IQR_outliers, np.zeros_like(IQR_outliers), s=5,color='red')
plt.yticks([])  
plt.title('IQR method')
plt.show()
```
