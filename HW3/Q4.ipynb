```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
binomial=scipy.stats.binom(n=10,p=0.5)
prob=[]
for i in range(0,11):
    prob.append(binomial.pmf(k=i))

indexes=np.arange(0,11)
plt.bar(indexes,prob)
plt.xticks(np.arange(0,11,step=1))
plt.show()


#--------------------------------------------------
print('------------------------------------------------------------------------------------')
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt


data=np.random.normal(loc=0,scale=1,size=1000)


plt.hist(data,bins=30,density=True)


x = np.linspace(-4, 4, 1000)
pdf_values=scipy.stats.norm.pdf(x,loc=0,scale=1)

plt.plot(x,pdf_values)

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
