```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randn(100)
y=3*x+0.5*np.random.randn(100)

print(np.cov(x,y))
print('--------------------------')
print(np.corrcoef(x,y))

plt.scatter(x,np.zeros_like(x),s=5)
plt.yticks([])
plt.show()

plt.scatter(y,np.zeros_like(y),s=5)
plt.yticks([])
plt.show()

#------------------------------------------------

x=np.random.randn(100)
y=-3*x+0.5*np.random.randn(100)

print(np.cov(x,y))
print('--------------------------')
print(np.corrcoef(x,y))

#if x and y have inverse relation,elements of Subdiameter on cov and correlation are negative


plt.scatter(x,np.zeros_like(x),s=5,color='red')
plt.yticks([])
plt.show()

plt.scatter(y,np.zeros_like(y),s=5,color='red')
plt.yticks([])
plt.show()
```
